# 2 - COMPUTER VISION - Color Spaces Project
    The aim of this task is to prepare a function to perform threshold selecting using HLS color space

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
image = mpimg.imread('images/test6.jpg') 
plt.imshow(image)
```

```python
def hls_selection(img, thresh=(0, 255)):
    hls_image = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    S = hls_image[:,:,2]
    binary_output = np.zeros_like(S)
    binary_output[(S > thresh[0]) & (S <= thresh[1])] = 1
    return binary_output
```

```python
hls_binary = hls_selection(image, thresh=(90, 255))
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()

ax1.imshow(image)
ax1.set_title('Original Image', fontsize=50)

ax2.imshow(hls_binary, cmap='gray')
ax2.set_title('Thresholded S', fontsize=50)

plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
